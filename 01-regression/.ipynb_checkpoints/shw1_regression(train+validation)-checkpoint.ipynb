```python
import pandas as pd
import numpy as np
data = pd.read_csv('/Users/shuiguangshan/Mydocument/Grade 1 master/06 ML/02 CodeEveryday/01-regression/train.csv', encoding = 'big5')
```

```python
data = data.iloc[:, 3:]
data[data == 'NR'] = 0
raw_data = data.to_numpy()
raw_data[:20, :]
```

```python
## 特征工程数据处理
month_data = {}
for month in range(12):
    sample = np.empty([18, 20 * 24])
    for day in range(20):
        sample[:,  day * 24 : (day + 1) * 24] = raw_data[18 * (month * 20 + day)  : 18 * (month * 20 + day + 1) , :]
    month_data[month] = sample
```

```python
## 取9个数据为一组
x = np.empty([12 * 471, 9 * 18], dtype = float)
y = np.empty([12 * 471, 1], dtype = float)
for month in range(12):
    for day in range(20):
        for hour in range(24):
            if (day == 19) & (hour > 14):
                continue
            x[471 * month + 24 * day + hour, :] = month_data[month][:, day * 24 + hour: day * 24 + hour + 9].reshape(1,-1)
            y[471 * month + 24 * day + hour, :] = month_data[month][9, day * 24 + hour + 9]
```

```python
## 标准化
mean_x = np.mean(x, axis = 0)
std_x = np.std(x, axis = 0)
for i in range(len(x)):
    for j in range(len(x[0])):
        x[i][j] = (x[i][j] - mean_x[j]) / std_x[j]
```

```python
## 构造验证集和训练集
import math
x_train_set = x[:math.floor(len(x) * 0.8), :]
y_train_set = y[:math.floor(len(y) * 0.8), :]
x_validation = x[math.floor(len(x) * 0.8):, :]
y_validation = y[math.floor(len(y) * 0.8):, :]
```

 ## train Model


```python
## 利用adagrad方法进行梯度优化
dim = 18 * 9 + 1
w = np.zeros([dim, 1])
x_train_set = np.concatenate((np.ones([len(x_train_set), 1]), x_train_set), axis = 1).astype('float')
learning_rate = 100
iterations = 100000
eps = 0.0000001
adagrad = np.zeros([dim, 1])
for iteration in range(iterations):
    loss_train_set = np.sqrt(np.sum(np.power(np.dot(x_train_set, w) - y_train_set, 2)) / len(x_train_set))
    if (iteration%10000 == 0):
        print(str(iteration) + ':' + str(loss_train_set))
    gradient = 2 * np.dot(x_train_set.transpose(),np.dot(x_train_set, w)- y_train_set)
    adagrad += gradient ** 2
    w -= learning_rate * gradient / (adagrad + eps)
w
```

```python
x_validation = np.concatenate((np.ones([len(x_validation),1]), x_validation), axis = 1)
loss_validation = np.sqrt(np.sum(np.power(np.dot(x_validation, w) - y_validation, 2)) / len(x_validation))
loss_validation
```
